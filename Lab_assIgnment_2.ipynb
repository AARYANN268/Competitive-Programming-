```python
# Input two positive integers
A = int(input("Enter first number: "))
B = int(input("Enter second number: "))

# -----------------------------
# Method 1: Basic Method
# -----------------------------
gcd_basic = 1
for i in range(1, min(A, B) + 1):
    if A % i == 0 and B % i == 0:
        gcd_basic = i

print("GCD using Basic Method:", gcd_basic)

# -----------------------------
# Method 2: Euclidean Algorithm
# -----------------------------
a, b = A, B
while b != 0:
    a, b = b, a % b

print("GCD using Euclidean Algorithm:", a)
```

```python
# Input a positive integer
N = int(input("Enter a number: "))

count = 0

# Loop from 1 to N
for i in range(1, N + 1):
    if N % i == 0:
        count += 1

print("Total number of divisors:", count)
```

```python
# Input
N = input("Enter N: ")
K = int(input("Enter K: "))

# Step 1: Concatenate N exactly K times
num = N * K

# Step 2: Repeatedly multiply digits until single digit
while len(num) > 1:
    product = 1
    for digit in num:
        product *= int(digit)
    num = str(product)

# Output
print(int(num))
```
